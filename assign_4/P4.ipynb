```python
Q1=df['medv'].quantile(0.25)
Q3=df['medv'].quantile(0.75)
IQR=Q3-Q1
ll=Q1-1.5*IQR
ul=Q3+1.5*IQR
print(f'Q1={Q1},Q3={Q3},iQR={IQR},Upper limit={ul},Lower Limit ={ll}')
```

```python
outliers=[]
for i in df.medv:
    if i<ll or i>ul:
        outliers.append(i)

print('\n Outliers are : ',outliers)
```

```python
df[df.medv < ll].index
```

```python
X = df.drop(['medv'],axis=1)  # All features except the target variable
Y = df['medv']                 # Target variable
```

```python
X
```

```python
df[df.medv > ul].index
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
import sklearn
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
```

```python
model=lm.fit(xtrain,ytrain)
```

model()


```python
model
```

```python
ytrain_pred=lm.predict(xtrain)
ytest_pred=lm.predict(xtest)
ytrain_pred
```

```python
ytrain
```

```python
import numpy as np
import sklearn.metrics
from sklearn.metrics import mean_absolute_error
```

```python
mse = sklearn.metrics.mean_squared_error(ytrain, ytrain_pred)
mae = mean_absolute_error(ytrain, ytrain_pred)
rmse = np.sqrt(mse)
r2 = sklearn.metrics.r2_score(ytrain, ytrain_pred)
```

```python
print("The model performance for training set")

```

```python
mse = sklearn.metrics.mean_squared_error(ytest, ytest_pred)
mae = mean_absolute_error(ytest, ytest_pred)
rmse = np.sqrt(mse)
r2 = sklearn.metrics.r2_score(ytest, ytest_pred)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))

plt.scatter(ytrain, ytrain_pred, c='blue', marker='o', label='Training data')
plt.scatter(ytest, ytest_pred, c='lightgreen', marker='s', label='Test data')

# Perfect prediction reference line
min_val = min(ytrain.min(), ytest.min())
max_val = max(ytrain.max(), ytest.max())
plt.plot([min_val, max_val], [min_val, max_val], 'r--', label='Perfect prediction')

plt.xlabel('True Values')
plt.ylabel('Predicted Values')
plt.title("True vs Predicted Values")
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```
