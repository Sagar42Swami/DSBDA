```python
import seaborn as sns
import matplotlib.pyplot as plt
titanic = sns.load_dataset("titanic")
sns.barplot(x='pclass', y='age', data=titanic)
plt.show()
```

```python
sns.distplot(titanic['age'])
plt.show()
```

```python
sns.histplot(titanic['age'],bins=5)
plt.show()
```

```python
titanic['sex'].value_counts().plot(kind="pie", autopct="%.2f")
```

```python
sns.barplot(x='pclass', y='age', hue='sex', data=titanic)
plt.show()

```

```python
sns.boxplot(x='sex', y='age', hue='survived', data=titanic)
plt.show()
```

```python
sns.distplot(titanic[titanic['survived'] == 0]['age'], hist=False, color="red")
sns.distplot(titanic[titanic['survived'] == 1]['age'], hist=False, color="green")
plt.show()
```

```python
import pandas as pd
sns.heatmap(pd.crosstab(titanic['pclass'], titanic['survived']))
```
